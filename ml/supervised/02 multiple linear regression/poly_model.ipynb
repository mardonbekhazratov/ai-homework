# Data Splitting

## Train Split / Test Split

Generalisation Loss

```python
df = pd.read_csv('data/advertising.csv')
X = df.drop(columns=['Sales']).values
y = df['Sales'].values
```

```python
X.shape
```

```python
train_size = int(0.7 * X.shape[0])
train_size
```

```python
X_train = X[:train_size]
X_test = X[train_size:]

y_train, y_test = y[:train_size], y[train_size:]

X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
ix = np.random.permutation(X.shape[0])
```

```python
X[ix] # shuffled X
y[ix] # shuffled y

train_size = int(0.7 * X.shape[0])

X_train = X[ix][:train_size]
X_test = X[ix][train_size:]

y_train, y_test = y[ix][:train_size], y[ix][train_size:]

X_train.shape, X_test.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
std_scl = StandardScaler()
make_poly = PolynomialFeatures(degree=2, include_bias=False, interaction_only=True)
lin_reg = LinearRegression()
```

```python
X_train = std_scl.fit_transform(X_train)
X_train = make_poly.fit_transform(X_train)
```

```python
lin_reg.fit(X_train, y_train)
```

```python
lin_reg.score(X_train, y_train)
```

```python
from sklearn.metrics import root_mean_squared_error
```

```python
y_pred = lin_reg.predict(X_train)
```

```python
root_mean_squared_error(y_train, y_pred)
```

```python
X_test = std_scl.transform(X_test)
X_test = make_poly.transform(X_test)
```

```python
lin_reg.score(X_test, y_test)
```
